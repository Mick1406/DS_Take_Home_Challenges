# <center> Funnel Analysis: Take Home Challenge

## Challenge Description
You are looking at data from an e-commerce website. The site is very simple and has just 4
pages:
- The first page is the home page. When you come to the site for the first time, you can
only land on the home page as a first page.
- From the home page, the user can perform a search and land on the search page.
- From the search page, if the user clicks on a product, she will get to the payment page,
where she is asked to provide payment information in order to buy that product.
- If she does decide to buy, she ends up on the confirmation page
<br> <br>
The company CEO isn't very happy with the volume of sales and, especially, o sales coming
from new users. Therefore, she asked you to investigate whether there is something wrong in
the conversion funnel or, in general, if you could suggest how conversion rate can be improved.
Specifically, she is interested in :
- A full picture of funnel conversion rate for both desktop and mobile
- Some insights on what the product team should focus on in order to improve
conversion rate as well as anything you might discover that could help improve

```python
# Import required libraries
import pandas as pd
from functools import partial, reduce
import matplotlib.pyplot as plt

import IPython.display
```

## Import data

```python
# Read csv files in dfs
user = pd.read_csv('Funnel/user_table.csv')
home_page = pd.read_csv('Funnel/home_page_table.csv')
search_page = pd.read_csv('Funnel/search_page_table.csv')
payment_page = pd.read_csv('Funnel/payment_page_table.csv')
payment_confirmation = pd.read_csv('Funnel/payment_confirmation_table.csv')
```

```python
# Put all the data together in a single df 
dfs = [user, home_page, search_page,payment_page,payment_confirmation]

user_all = partial(pd.merge, on=['user_id'], how='outer')
user_all = reduce(user_all, dfs)

user_all.columns = ['user_id','date','device','sex',
                    'home_page','search_page','payment_page','payment_confirmation']
user_all.head()
```

## Data transformation and initial exploration

### Create the funnel field and related indicators

```python
# Funnel
lst_pages = ['home_page','search_page','payment_page','payment_confirmation']

user_all['funnel'] = user_all[lst_pages].astype(str).agg('->'.join, axis=1)
user_all['funnel'].value_counts()
```

```python
# Number of pages landed on
user_all['n_pages'] = 4 - user_all['funnel'].str.count('nan')
user_all['n_pages'].value_counts()
```

```python
# Date distribution 
user_date = user_all.groupby([user_all['date']])['user_id'].count()
display(user_all['date'].min())
display(user_all['date'].max())
user_date.plot()
```

```python
# Create a month variable
def pull_mth(string):
    return string[5:7]

user_all['month'] = user_all['date'].apply(pull_mth)

# Chart funnel by month
user_all['month'].value_counts()
```

## Analysis of funnel by month and device

```python
# Funnel distribution by month
# user_mth = user_all.groupby(['month','funnel']).agg({'user_id':'count'})
# user_mth = user_mth.reset_index()
user_mth_funnel = pd.crosstab(user_all['month'],user_all['funnel'],normalize='index')

plt.figure(figsize=(10,7))
user_mth_funnel.plot(kind='bar', stacked=True)
plt.legend(bbox_to_anchor=(1.1, 1.05))
```

```python
# Funnel distribution by device
user_device_funnel = pd.crosstab(user_all['device'],user_all['funnel'],normalize='index')

plt.figure(figsize=(10,7))
user_device_funnel.plot(kind='bar', stacked=True)
plt.legend(bbox_to_anchor=(1.1, 1.05))
```

```python
# Funnel distribution by monht for each device
desktop = user_all['device']=='Desktop'
mobile = user_all['device']=='Mobile'
user_mth_funnel_desktop = pd.crosstab(user_all[desktop]['month'],user_all['funnel'],normalize='index')
user_mth_funnel_mobile = pd.crosstab(user_all[mobile]['month'],user_all['funnel'],normalize='index')

plt.figure(figsize=(10,7))

# Desktop
display(user_mth_funnel_desktop.plot(kind='bar', stacked=True))
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.title('Funnel by month for Desktop device')

# Mobile
display(user_mth_funnel_mobile.plot(kind='bar', stacked=True))
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.title('Funnel by month for Mobile device')
```

## <font color='blue'> Key Insights:
    - Spike in traffic in Feb (in terms of visit, regardless of conversion)
    - In recent months (March and April), the conversion rate is much lower! Indeed, visitors of the e-commerce website don't go further than the home page
    - In those 2 months, visitors coming from mobile are much more likely to not go further than the home page
    - There may be an issue with mobile optimisation (maybe some changes to the site in March?)
